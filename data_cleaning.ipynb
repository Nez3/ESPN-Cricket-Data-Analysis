# ESPN Cricket (Test Match) Data (Data Cleaning in Python)

```python
# importing necessary packages
import numpy as np
import pandas as pd
```

```python
# reading the excel file in a dataframe using pandas
df = pd.read_csv('ESPN_cricket_data.csv')
df.head()
```

```python
# renaming columns to make them more readable
df = df.rename(columns= {'Span': 'playing_span','Mat': 'matches_played','NO': 'not_outs','HS': 'highest_innings_score','Ave': 'avg_batting', 'BF': 'balls_faced', 'SR': 'strike_rate', '100': 'hundreds_scored', '50': 'fifties_scored', '0': 'ducks_scored'})
```
